## Interpretation

Data is presented as a set of text files. Each file consists of lines like this:

    24.3892672440343 56.0953146612789 dev1 down 16 0 -79 -93 -62 -69 -87
    
**Note that the description of the data on CRAWDAD doesn't match the actual content of the files.** So already from this point on, we are guessing as to what the numbers in files actually mean.

Based on the `eWINE/linkQualityEstimation/datasets/trace2_Colorado/pythonScript/perLink.py`, our interpretation of the data is as follows:

Each line describes one packet, transmitted by a node and received by 5 "monitors". Columns in the files:

 * 0-1: Some kind of coordinate of the transmitting node.
 * 2: Unknown. Maybe wireless network device for nodes that have more than one.
 * 3: Orientation of the antenna.
 * 4: Transmit power in dBm.
 * 5: Packet sequence number.
 * 6-10: RSSI in dBm for each of the 5 monitors.

```python
class CURow():
    def __init__(self, line):
        fi = line.split()
    
        self.x = float(fi[0])    
        self.y = float(fi[1])

        self.dev = fi[2]
        self.ant = fi[3]

        self.pw = int(fi[4])

        self.seq = int(fi[5])

        self.rssi = map(float, fi[6:])
```

The `perLink.py` script treats each line in the file as information about 5 packets, sent over 5 "links". Each link is identified using the transmitting node, transmission power and received monitor.

```python
import glob
from collections import defaultdict

packets_in_link = defaultdict(list)

for dataset in glob.glob("/home/avian/dev/eWINE/linkQualityEstimation/datasets/trace2_Colorado/data/*.txt"):
    for line in open(dataset):
        row = CURow(line)
        
        for i, rssi in enumerate(row.rssi):
            link_id = (dataset, row.x, row.y, row.dev, row.ant, row.pw, i)
            
            packets_in_link[link_id].append(row)
```

For each link, a set number of packets is assumed to be sent. `perLink.py` gets this number from the file from which the data is read.

```python
expected_n_packets = {}
for link_id in packets_in_link.keys():
    
    dataset = link_id[0]
    if 'omni_16dbm' in dataset:
        n = 500
    elif 'omni_variable_txpower' in dataset:
        n = 200
    elif 'dir_variable_txpower' in dataset:
        n = 200
    else:
        assert False
        
    expected_n_packets[link_id] = n
```

The dataset has some rows missing, which is interpreted as lost packets. PRR for each link is calculated by dividing the expected number of packets on the link and the number of packets that is present in the file.

**Note: dataset description on the CRAWDAD doesn't mention anything about lost packets. It is very unusual that lost packets would be represented as a whole missing line. What about cases where just one of the 5 monitors did not receive the packet?**

Average RSSI for each link is calculated by calculating the mean over the monitor's RSSI.

```python
n_links = len(packets_in_link)

mean_rssi = empty(n_links)
prr = empty(n_links)

for i, (link_id, rows) in enumerate(packets_in_link.items()):
    
    n_received = len(rows)
    n_sent = expected_n_packets[link_id]
    
    prr[i] = float(n_received)/n_sent
    
    monitor = link_id[6]
    mean_rssi[i] = sum(row.rssi[monitor] for row in rows)/float(n_received)
```

Plotting average RSSI versus PRR for all links:

```python
plot(mean_rssi, prr, '.')
xlabel("Mean RSSI [dBm]")
ylabel("PRR")
grid()
```
